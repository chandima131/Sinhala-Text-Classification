```python
#!/usr/bin/env python
# -*- coding:utf-8 -*-

import codecs
data = pd.read_csv('Single label dataset for preprocessing 2.csv', header = None, encoding='utf-8', engine='python')
data.columns = ['ID','User','FULL_TEXT','CREATED_AT','LIKE COUNT','FOLLOWERS COUNT','RETWEET COUNT','LABEL']
# data = data.dropna()
data = data.iloc[1: , :]
data.head()
```

```python
def clean_text(text):
    tokens = nltk.word_tokenize(text)
    regex = re.compile(u'[^\u0D80-\u0DFF]', re.UNICODE)
    tokens = [regex.sub('', w) for w in tokens]
    tokens = [c for c in tokens if c not in string.punctuation]
    return tokens
```

```python
data['cleaned_msg'] = data['FULL_TEXT'].apply(lambda x: clean_text(x))
data.head()
```

```python
############################ Stop Word removing Method 2 - using predefined stop words ####################################
```

```python
stopwordsreader = open('StopWords', 'r', encoding='utf-8')
reader = csv.reader(stopwordsreader)

swords = []
for row in reader:
    swords.append(row[0])
print(swords)
    
```

```python
def remove_stop_words(tokens):
    regex = re.compile(u'[^\u0D80-\u0DFF]', re.UNICODE)
    tokens = [regex.sub('', w) for w in tokens]
    tokens = [c for c in tokens if c not in string.punctuation]
    return tokens
```

```python
data['remove_stopword'] = data['cleaned_msg'].apply(lambda x: remove_stop_words(x))
data.head()
```

```python
########################################## Stemming ##################################################
```

```python
stem_dict = open("stem_dictionary.txt", "r", encoding='utf-8')
stem_dict = map(lambda s: s.strip(), stem_dict)

stem_dictionary = {}

for s in stem_dict:
    s = s.split("\t")
    stem_dictionary[s[0]] = s[1]
    
```

```python
def stemming_words(tokens):
    for k, v in enumerate(tokens):
        tokens[k] = stem_dictionary.get(v, v)
    tokens = " ".join([c for c in tokens if c not in string.punctuation])
    return tokens
```

```python
data['STEMMING WORDS'] = data['remove_stopword'].apply(lambda x: stemming_words(x))
data.head()
```

```python
######################### Remove Duplicates from dataset #################
```

```python
data.drop_duplicates(subset='STEMMING WORDS',keep='first',inplace = True)
data.head()
```

```python
df = pd.DataFrame(data, columns= ['User', 'CREATED_AT','LABEL','STEMMING WORDS','LIKE COUNT','FOLLOWERS COUNT','RETWEET COUNT'])
df.to_csv("FinalListPreprocessedDataNew.csv")
```
